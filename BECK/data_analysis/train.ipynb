# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importación dataset

```python
df = pd.read_csv(
    '../datasets/dataset_entrenamiento_coseno.csv', encoding='utf-8')
```

#### Información

```python
# Primeros 10 datos
print(df.head(10))
```

```python
# Info del dataset
print(df.info())
```

```python
# Describir el dataset
print(df.describe())
```

```python
# Numero de datos por clase
#* 4070 -> Clase 1 -> Depresión
#* 2996 -> Clase 0 -> No depresión
print(df['Clase'].value_counts())
```

```python
# Distribución de las clases
%matplotlib inline
df['Clase'].hist()
```

```python
# Representar gráficamente la distribución de los atributos
df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# Mostrar la correlación entre los atributos del conjunto de datos
corr_matrix = df.corr()
corr_matrix["Clase"].sort_values(ascending=False)
```

```python
# Mostrar correlación lineal entre todos los atributos del conjunto de datos
df.corr()
```

```python
# Representar gráficamente la matriz de correlación
corr = df.corr()
fig, ax = plt.subplots(figsize=(8, 8))
ax.matshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
```
